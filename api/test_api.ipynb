```python
from pprint import pprint as pp
import requests
```

```python
def view_pretty(text):
    text = text.replace('true', 'True')
    text = text.replace('false', 'False')
    text = text.replace('null', 'None')
    
    pp(eval(text))

url = 'http://127.0.0.1:8000/api'
```

```python
data = {
    'kind': 'Собака',
    'sex': True,
    'min_weight': 5,
    'max_weight': 60,
    'min_age': 0
}
```

```python
view_pretty(requests.get(url, {'section': 'animal', 'method' : 'filter', 'params': str(data)}).text)
```
